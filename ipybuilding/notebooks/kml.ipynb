```python
import geopandas
import fiona
fiona.drvsupport.supported_drivers['LIBKML'] = 'rw' 
states = geopandas.read_file('./data/3D_geneve_KML.kmz')
```

```python
states['geometry']
```

```python
import pydeck

INITIAL_VIEW_STATE = pydeck.ViewState(
  latitude=46.24,
  longitude=6.2,
  zoom=11,
  max_zoom=16,
  pitch=45,
  bearing=0
)


geopandas_layer= pydeck.Layer('GeoJsonLayer', data=states, get_fill_color=[255,255,255], pickable=True,)

r = pydeck.Deck(
    layers=[geopandas_layer],
    initial_view_state=INITIAL_VIEW_STATE)

# r.show()
r.to_html()
```
